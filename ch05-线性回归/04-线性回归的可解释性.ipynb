# 对数据的可解释性

先使用线性的方式，试试看

看那个特征，影响比较大！没坏处


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
boston = datasets.load_boston()
X = boston.data
y = boston.target

X = X[y < 50.0]
y = y[y < 50.0]
```

```python
X.shape
```

```python
X.shape
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
lin_reg.coef_
```

```python
# 排序索引， 从小到大 
np.argsort(lin_reg.coef_)
```

系数越小，对房价影响越负面，负相关

系数越大，对房价的影响越明显

```python
boston.feature_names[np.argsort(lin_reg.coef_)]
```
