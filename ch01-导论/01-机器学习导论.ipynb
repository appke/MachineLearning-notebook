## 机器学习导论

我来回答小哥对谷歌提出的第一个问题。

- 机器学习(AI的一个分支)需要尽可能大的数据集合(称为Knowledge base, 或语料库Corpus, 或dataset)让机器利用它们训练(Training, 相应的数据集合称为Training Set)出一个设计好的算法方法很多
    - 例如神经网络, 
    - 深度神经网络, 
    - 支持矢量机SVM, 
    - 暗含马尔可夫模型HMM)的参数。
- 参数是组成算法的必要基础, 参数必须经过训练和测试(test), 才能确定, 这就是基本的机器学习的概念。
- 如果(大)数据集合具有偏向性, 而不全面, 比如人脸识别的数据集合主要由雅利安白人的脸型组成, 或者缺乏足够的中文语言例句。
    - 于是训练出来的任何算法都具有极大的偏见性, 
    - 谷歌可以正确识别白种人的人脸, 却在识别中国人脸的结果里出现大误差。
- 同样的, 识别极其复杂的汉语方言语言上进展不会大。
    - 因此, 谷歌不惜与中共当局勾结, 由李飞飞(我相当怀疑她的研究水平, 尽管她在斯坦福教书)建立一个中国中心, 
        - 以获取尽可能多的中国特有而又极其丰富的数据。
    - 中国庞大的人口及相应庞大的上网人口, 是一个宝贵的数据资源, 
        - 这是任何进行人脸图像识别和自然语言分析的公司和研究机构垂涎三尺的。

- 实际上, zg网络监管利用大数据和人工智能(比方说图像识别和语音识别)还做不到监控每个人, 网上每一"有害"信息的精确监控。
- 只能做到重点监控(即建立一个黑名单, 对黑名单里列入的人进行实时监控)。
    - 不但中共虚夸而自吹自擂的技术公司做不到, 就是美国先进公司都做不到!  
    - 要达到精确跟踪每一个人的行迹, 必须做到数据共享。
        - 例如, 若某人在广西街上违规过街, 他的面貌被摄像头摄下, 就算机器能识别出他的脸, 然而他的人脸数据是储备在山西省个人资料数据仓库里的, 要想确定他的身份, 就必须通过网络传输去山西省数据仓库里取来对比。如果只有几百个人则数据传输没有多大问题, 但若每天每时几十万, 几百万, 传输就会出现堵塞! 所以除了重点监控的黑名单上的人以外, 事实上做不到对每一个人自动监控的。 
    - 其次, 识别本身的误差极大, 在机器学习里称为"模糊性", 特别是在噪音干扰下, 目前识别率很低, <40%。
    - 语音识别差错就更大了。所以小哥不必变音来视频。
        - 例如人在某处犯罪, 只要他不是惯犯上了黑名单, 在光线模糊, 如夜晚或黄昏时摄入镜头被记录下的图像基本不能被识别, 老美做不到, 老中更做不到! 
        - 再例, 关于微信群里的关键词识别问题。这个也只是对大群或被网管盯梢了的群做得到, 对几千万微信群是无法筛选出来的, 因为这将极大减慢网速。其次可以把代码文字转化为图像文字。网络长城只能靠一种称为OCR技术先将图像文字转化为代码文字, 再对代码文字内容进行敏感词过滤。
        - 但是, 若图像清晰度不高, 上面被胡乱化了线, 或额外加了背景花色, 则OCR无法正确转化! 否则只能用人眼一个个盯着看。
