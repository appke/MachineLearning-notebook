## 一维数组
> 和python数组不一样，只能是同一类型

```python
import numpy as np
np.__version__
```

```python
a = np.arange(0, 10)
b = np.arange(10)
c = np.arange(0,10,2)
print(a, b, c)
```

```python
# 将一维数组转化为 二维数组
X = np.arange(15).reshape(3, 5)
X
```

```python
X[2,3]
```

#### 数组切片
用元祖，单括号

```python
# 前2行， 前3列
X[:2, :3]
```

```python
# 前2行， 前3列
X[:2]
```

```python
X[:, :3]
```

```python
# 前2行 间隔为2；切片表达式放在不同的维度中
X[:2, ::2]
```

```python
# 矩阵反转
X[::-1, ::-1]
```

```python
X[0, :]
```

```python
X[:, 1].ndim
```

```python
# 子矩阵，会影响原矩阵
# 和Python list 不同，优先考虑效率
subX = X[:2, :3]
subX[0, 0] = 111
X
```

```python
subX = X[:2, :3].copy()
subX[0, 0] = 112
X
```
